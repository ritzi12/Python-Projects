## Load Packages

```python
#import libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy.random import seed
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

## Load Dataset

```python
iris_df=pd.read_csv('Iris.csv',header='infer',index_col=0)

iris_df.describe(include='all')
```

## Exploratory Data Analysis

We check correlation matrix among features and observe that 'Petal Length' and 'Petal Width ' have a high positive correlation of 0.96.

Also the pairplots of seaborn allow us to observe distribution of features corresponding to various classes. The 'Iris- sentosa' class is easily distinguishable in all feature variables but there is overlapping in 'Iris-virginica' and 'Iris-versicolor.' 

```python
#Correlation Matrix Heat Map 
sns.heatmap(iris_df.corr(), annot=True,cmap='plasma')
```

```python
palette={'Iris-virginica':'chartreuse','Iris-setosa':'magenta','Iris-versicolor':'orangered'}
sns.pairplot(iris_df,hue='Species',palette=palette)
```

## Splitting and seperating Dataset

We sperate feature and target variable .And spplit dataset into test and train.We use 25% of dataset as test data.

```python
#Seperate Feature and Targets
X=iris_df.iloc[:,0:4]
Y= iris_df['Species']

#Train Test Split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=3)
```

## Building Decision Tree Model

```python
#Building Decision Tree
model=DecisionTreeClassifier()

seed(100)

model=model.fit(X_train,Y_train)
```

## Predict Test Data

```python
#Predict test Data 
test_pred=model.predict(X_test)

test_pred
```

## Evaluate Model on Test Data

We evaluate the test data predictions and acheive accuracy of 95% on test data set i.e. our model predicts correct class for petals for 95% of test data.

```python
#Evaluate Model
print("Accuracy:",metrics.accuracy_score(Y_test, test_pred))
```

```python
#Results
df={'Test_Label':Y_test,'Pred_Label':test_pred}
pd.DataFrame(df)
```

## Visualize Decision Tree

```python
class_names=iris_df['Species'].unique()
```

```python
#Visualize Decision Tree

from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(model, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names = X.columns,class_names=class_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('petals.png')
Image(graph.create_png())
```
